#딕셔너리 구조의 막대 그래프

````python
lstColor = ['pink','blue','#00ffff','r']
dict1 = {'index':['사과','바나나','키위','멜론'],
         'data' : [30,20,40,18]}
plt.bar('index','data',data=dict1, color=lstColor)
plt.show()
````

````python
# 원그래프
# explode : 조각 분리, shadow :그림자
# wedgeprops(linewidth : 테두리 두께, edgecolor :'테두리색')
plt.title('과일선호도조사')
lstColor = ['pink','blue','#00ffff','r']
dict1 = {'index':['사과','바나나','키위','멜론'],
         'data' : [30,20,40,18]}
plt.pie('data',labels='index',data=dict1, colors=lstColor, startangle=90,autopct='%.1f%%',
        explode=(0.01, 0.3, 0.01, 0.01), shadow=True, wedgeprops={'linewidth':1 , 'edgecolor':'blue'})
plt.show()
````

#데이터 전처리

````python
df.columns
df = df.rename(columns={'발생년':'년도'})
df
````

````python
#특정 열의 데이터 형을 변경하기
df['년도'] = df['년도'].astype('str')
df.info()
````

````python
# 필요하지 않은 열 제거
df.drop(columns=['중상자수','경상자수','부상신고자수'])
df
````

````python
# 필요하지 않은 열 제거
df.drop(columns=['중상자수','경상자수','부상신고자수'], inplace=True)
df
````

````python
# 시각화하기
# 년도별로 사고 건수 시각화
plt.rcParams['font.size']=12
plt.rcParams['lines.linewidth']=2
plt.figure(figsize=(6,3))
plt.title('년도별 사고 건수', fontsize=20)
plt.plot(df['년도'],df['사고건수'], c='b' ,marker='^', label = ' 사고건수')
plt.legend() #범례
plt.grid() #눈금선
plt.show()
````

````python
# 시각화하기
# 년도별로 사망자 사고 건수 시각화
plt.rcParams['font.size']=12
plt.rcParams['lines.linewidth']=2
plt.figure(figsize=(6,3))
plt.title('년도별 사망자 사고 건수', fontsize=20)
plt.plot(df['년도'],df['사망자수'], c='r' ,marker='o', label = ' 사망자수')
plt.legend() #범례
plt.grid() #눈금선
plt.show()
````

````python
# 시각화하기
# 년도별로 사망자 사고 건수 시각화
plt.rcParams['font.size']=12
plt.rcParams['lines.linewidth']=2
plt.figure(figsize=(6,3))
plt.title('년도별 사망자 사고 건수', fontsize=20)
plt.plot(df['년도'],df['사망자수'], c='r' ,marker='o', label = ' 사망자수')
plt.plot(df['년도'],df['사고건수'], c='b' ,marker='^', label = ' 사고건수')
plt.legend() #범례
plt.grid() #눈금선
plt.show()
````

````python
#두 그래프를 겹쳐 그리기
fig, ax1 = plt.subplots()
ax1.bar(df['년도'],df['사고건수'], color='b' , label = ' 사고건수')
ax1.set_ylim(150000,240000)

ax2 =ax1.twinx()
# "maker" should be "marker"
ax2.plot(df['년도'], df['사망자수'], c='r' , marker='o', label = ' 사망자수')
ax2.set_ylim(2500, 4000)

plt.suptitle('년도별 사고건수와 사망자수의 관계')

# Call plt.show() to display the plot
plt.show()
````
